```python
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure

import pandas as pd
```

```python
# .py script
df_uk = pd.read_csv("ukData/results/200uk.csv")

# ipynb run
df_uk_n = pd.read_csv("ukData/results/uk200_note_norm.csv")

# .py script sans CHESS data
df_uk_noCHESS = pd.read_csv("ukData/results/200uk_noCHESS.csv")
```

```python
for key in df_uk_noCHESS.keys():
    print(key)
    plt.scatter(df_uk_noCHESS.simDay, df_uk_noCHESS[key])
    plt.show()
```

```python
def plot_it(df):
    df_plot = df.copy(deep=True)

    figure(num=None, figsize=(20, 10), dpi=80, facecolor='w', edgecolor='k')

    #Set labels
    if df["susceptible"][0] == 1:
        ylabel='Percent Population in Health State'
        title = 'Normalized Population: State Change Over Time'  
    else:
        ylabel= 'Population in Health State'
        title = 'Total Population: State Change Over Time'

    states = ["susceptible", "exposed", "asymptomatic", "infected1", "infected2", "recovered1", "recovered2", "deceased"]
    
    df_plot.plot(x='simDay', y=states, 
                 xlabel='Simulation Day',
                 ylabel= ylabel,
                 title = title,
                 figsize=(15,10))
    plt.show()
```

```python
def plot_et(df):
    df_plot = df.copy(deep=True)

    figure(num=None, figsize=(20, 10), dpi=80, facecolor='w', edgecolor='k')
    ylabel= 'Population in Health State'
    title = 'Total Population: State Change Over Time'

    states = ["S", "E", "A", "I1", "I2", "R1", "R2", "D"]
    
    df_plot.plot(x='simDay', y=states, 
                 xlabel='Simulation Day',
                 ylabel= ylabel,
                 title = title,
                 figsize=(15,10))
    plt.show()
```

```python
plot_it(df_uk_noCHESS)
```
